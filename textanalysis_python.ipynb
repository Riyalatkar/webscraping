```python
'''code to make stop words list given in tex files'''
path = r"C:\Users\Lenovo\StopWords"

# Change the directory
os.chdir(path)

# Read text File
stopwords=[]

def read_text_file(file_path):
    with open(file_path, 'r') as f:
        for words in f:
            stopwords.append(words.strip())


# iterate through all file
for file in os.listdir():
# Check whether file is in text format or not
    if file.endswith(".txt"):
        file_path = f"{path}\{file}"

# call read text file function
        read_text_file(file_path)
```

```python
pos=r"C:\Users\Lenovo\Downloads\positive-words.docx"
neg=r"C:\Users\Lenovo\Downloads\negative-words.docx"
positive_words=[]
negative_words=[]
# create an instance of a 
# word document we want to open
def conc(url,listi):
    doc = Document(url)
    for para in doc.paragraphs:

        if (para.text =="" or para.text==" "):
            continue
        listi.append(para.text)
    return listi
conc(pos,positive_words)
conc(neg,negative_words)
#print(positive_words)
```

```python
for url in allurls:
    article=Article(url)
    try:
        article.download()
        article.parse()
    except:
        continue
    
    article.nlp()
    
    text=article.text
    test_str = text.translate(str.maketrans('', '', string.punctuation))
    text_tokens = word_tokenize(test_str)
    tokens_without_sw = [word.lower() for word in text_tokens if not word in stopwords]
    def pnps(tokens_without_sw):
        positive_score=0
        negative_score=0

        for word in tokens_without_sw:
            if word in positive_words:
                positive_score+=1
            if word in negative_words:
                negative_score+=1
            else:
                tokens_without_sw.remove(word)
        polarity_score=(positive_score - negative_score)/ (positive_score + negative_score + 0.000001)
        subjectivity_score = (positive_score + negative_score)/ ( len(tokens_without_sw)+ 0.000001)
        return (positive_score,negative_score,polarity_score,subjectivity_score)
    ans1234=pnps(tokens_without_sw)

    posiscore=ans1234[0]
    negiscore=ans1234[1]
    polarscore=ans1234[2]
    subjecscore=ans1234[3]

    def avg_sentence_len(text):
        sentencess = text.split(".") #split the text into a list of sentences.
        words = text.split(" ") #split the input text into a list of separate words
        if(sentencess[-1]==""): #if the last value in sentences is an empty string
            average_sentence_length = len(words) / len(sentencess)-1
        else:
            average_sentence_length = len(words) / len(sentencess)
        return average_sentence_length #returning avg length of sentence
    avgsenlen = avg_sentence_len(text) 

    def complex_words_percentage(text):
        # Tokenize the text into words
        words = nltk.word_tokenize(text)

        # Use NLTK's part-of-speech tagging to identify complex words
        tagged_words = nltk.pos_tag(words)
        complex_words = [word for word, tag in tagged_words if tag in ["JJ", "JJR", "JJS", "RB", "RBR", "RBS"]]

        # Calculate the percentage of complex words
        total_words = len(words)
        complex_words_count = len(complex_words)
        percentage = (complex_words_count / total_words) * 100

        return percentage,complex_words_count
    complexvocab = complex_words_percentage(text)
    cpercentage=complexvocab[0]
    lencomplex_words=complexvocab[1]

    fog_index=0.4 * (avgsenlen+ cpercentage)

    def Avg_words_persen(text):
        #text = "This is a sample text. It contains multiple sentences. We will use this text to calculate the average number of words per sentence."

    # Split the text into sentences using the '.' as the delimiter
        sentences = text.split('.')

        # Remove any empty strings from the list
        sentences = [sentence for sentence in sentences if sentence.strip() != '']

        # Count the number of words in each sentence
        word_counts = [len(sentence.split()) for sentence in sentences]

        # Calculate the average number of words per sentence
        average = sum(word_counts) / len(sentences)

        return average
    avgwordspersent=Avg_words_persen(text)

    wordslen = len(text.split(" "))
    #
    wordss = text.split(" ") 
    syllable_count=0
    for word in wordss:
        syllable_count +=textstat.syllable_count(word)

    SYLLABLE_PER_WORD=syllable_count/wordslen
    #
    personal_pronouns = re.findall(r'\b(I|me|you|he|him|she|her|it|we|us|they|them)\b', text)
    listp=len(personal_pronouns)
    
    avgwords = sum(len(word) for word in wordss) / len(wordss)
    
    df.loc[df['URL'] == url, 'POSITIVE SCORE'] = posiscore
    df.loc[df['URL'] == url, 'NEGATIVE SCORE'] = negiscore
    df.loc[df['URL'] == url, 'POLARITY SCORE'] = polarscore
    df.loc[df['URL'] == url, 'SUBJECTIVITY SCORE'] = subjecscore
    df.loc[df['URL'] == url, 'AVG SENTENCE LENGTH'] = avgsenlen
    df.loc[df['URL'] == url, 'PERCENTAGE OF COMPLEX WORDS'] = cpercentage
    df.loc[df['URL'] == url, 'FOG INDEX'] = fog_index
    df.loc[df['URL'] == url, 'AVG NUMBER OF WORDS PER SENTENCE'] = avgwordspersent
    df.loc[df['URL'] == url, 'COMPLEX WORD COUNT'] = lencomplex_words
    df.loc[df['URL'] == url, 'WORD COUNT'] = wordslen
    df.loc[df['URL'] == url, 'SYLLABLE PER WORD'] = SYLLABLE_PER_WORD
    df.loc[df['URL'] == url, 'PERSONAL PRONOUNS'] = listp
    df.loc[df['URL'] == url, 'AVG WORD LENGTH'] = avgwords
```
